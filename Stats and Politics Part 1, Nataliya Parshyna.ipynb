### In this report, I will analyze the data from previous United States presidential elections. USA elections are held every four years, and I will explore data for 2012 and 2016.

    An election for president of the United States occurs every four years on the first Tuesday after the first Monday in November. The president and vice president are elected through a process called the Electoral College which is a compromise between a popular vote by citizens and a vote in Congress.  Before the general election, most candidates for president go through a series of state primaries and caucuses which let the states choose the major political partiesâ€™ nominees for the general election.
    The president must:
    -be a natural-born citizen of the United States, 
    -be at least 35 years old, 
    -have been a resident of the United States for 14 years. 
    Anyone who meets these requirements can declare their candidacy for president. Once a candidate raises or spends more than $5,000 for their campaign, they must register with the Federal Election Commission.

## Part 2: Cleaning

### In the second part of this report, I will attempt to clean the data, i.e. to exclude unnecessary to the asked questions information, exclude redundant data, and also exclude null values if there are any.



   Before I start cleaning my data, I am going to perform another joint over 'votes' and 'demographics' data sets.

```python
# I need to join all tables togeter and check for duplicate columns. I already have a joined data frame for Votes.
# Next I will join Votes with Demographics.

result_Merged_info = pd.merge(df_Votes, df_Demographic_data, how='outer', on=['FIPS'] )
```

```python
result_Merged_info.head()
```

```python
result_Merged_info.info()
```

```python
# I will check if there are any null values

pd.DataFrame(result_Merged_info.isna().sum(), columns=["Number of Missing Values"])
```

```python
# I decided to delete columns that have redundant information: 
    
# - 'black_and_hispanic_pct' ( we have black and hispanic columns separately),  so there is no new information there
# - 'nonwhite_pct' - same reason
#- 'caucasian_frac' - same reason

result_Merged_info.drop(['black_and_hispanic_pct', 'nonwhite_pct', 'caucasian_frac', ], axis=1, inplace=True)
```

```python
result_Merged_info.head()
```

```python
# I will remove the null values where the 'Candidate' is 'other'

indexNames = result_Merged_info[(result_Merged_info ['candidate'] == 'Other')].index
```

```python
result_Merged_info.drop(indexNames, inplace=True)
```

```python
result_Merged_info.head()
```

```python
# I am checking for the percent of missing numbers

result_Merged_info.isna().sum()/result_Merged_info.shape[0]*100
```

    To answer my questions, I do not need to have columns: citizen_voting_age_population, foreign_born_pct, 
 age_29_and_under_pct, age_65_and_older_pct,unemployed_pct_in_civilian_labor_force,   less_than_highschool_pct,less_than_college_pct. I will drop the above columns.

```python
# Dropping columns

result_Merged_info.drop(['citizen_voting_age_population', 'foreign_born_pct',  
                         'age_29_and_under_pct', 'age_65_and_older_pct',
                         'unemployed_pct_in_civilian_labor_force', 'less_than_highschool_pct',
                         'less_than_college_pct'], axis=1, inplace=True)
```

```python
result_Merged_info.head()
```

```python
# Checking for the percent of missing numbers

result_Merged_info.isna().sum()/result_Merged_info.shape[0]*100
```

```python
# I will also drop the column county_y as its about 1 % od data missing - not significant

result_Merged_info.drop(['county_y'], axis=1, inplace=True)
```

```python
result_Merged_info.head()
```

```python
result_Merged_info.isna().sum()/result_Merged_info.shape[0]*100
```

   Now I am going to deal with the rest of the missing data. I will put missing values in all the columns that are missing values except the 'rural population_pct' by finding the mean of those columns.

```python
result_Merged_info = result_Merged_info.fillna(value=
                                              {'total_population':result_Merged_info.total_population.mean(),
                                               'white_pct':result_Merged_info.white_pct.mean(),
                                               'black_pct':result_Merged_info.black_pct.mean(),
                                                'hispanic_pct':result_Merged_info.hispanic_pct.mean(),
                                                'female_pct':result_Merged_info.female_pct.mean(),
                                                'median_household_income':result_Merged_info.median_household_income.mean()
                                             })
```

```python
result_Merged_info.head()
```

    And finally, I will remove all rows with null numbers from the column 'rural population_pct' (just to try a different method).
    

```python
result_Merged_info = result_Merged_info.dropna(subset=['rural population_pct'])
```

```python
result_Merged_info.head()
```

```python
result_Merged_info.isna().sum()/result_Merged_info.shape[0]*100  # no null values in our data set
```

#### As a result of cleaning data, I excluded all the null values, and now my data looks much cleaner and more transparent.  

```python
result_Merged_info.dtypes
```
